```python
import numpy as np
import pickle
import gym
import gc
import random
from torch.utils.tensorboard import SummaryWriter
import matplotlib.pyplot as plt
from typing import List, Tuple, Dict, Set, Union

gc.enable()
gc.collect()
```

```python
# CONFIG
H = 200 # num hidden nodes
N = 16 # num nodes visualized
```

```python
# VISUALIZE EACH NODE

def visualize_nodes(layer,
                    title: str = None):
    fig, axs = plt.subplots(2, 8, figsize=(16,4))

    for i in range(N):
        # j = random.randint(0,H)
        node = layer[i].reshape(80,80)

        x = i // 8
        y = i % 8

        axs[x,y].imshow(node, cmap='gray')

    fig.suptitle(title)
    plt.show()
```

```python
# LOAD MODELS

path = 'save_exp5_4000'
model = pickle.load(open(f'checkpoints/{path}.p','rb'))
layer = model['W1']
visualize_nodes(layer, 'Node Visualization for save_exp5_4000 (split batch training)')

# LOAD MODELS

path = 'save_7500'
model = pickle.load(open(f'checkpoints/{path}.p','rb'))
layer = model['W1']
visualize_nodes(layer, 'Node Visualization for save_7500 (vanilla)')

path = ''
```
