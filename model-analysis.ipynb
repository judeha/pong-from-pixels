```python
# CONFIG
H = 200 # num hidden nodes
N = 16 # num nodes visualized
```

```python
# VISUALIZE EACH NODE

def visualize_nodes(layer,
                    title: str = None):
    fig, axs = plt.subplots(2, 8, figsize=(16,4))

    for i in range(N):
        # j = random.randint(0,H)
        node = layer[i].reshape(80,80)

        x = i // 8
        y = i % 8

        axs[x,y].imshow(node, cmap='gray')

    fig.suptitle(title)
    plt.show()
```

```python
# LOAD MODELS

path = 'save_500'
model = pickle.load(open(f'checkpoints/{path}.p','rb'))
layer = model['W1']
visualize_nodes(layer, f'Node Visualization for {path} (vanilla)')

path = 'save_4000'
model = pickle.load(open(f'checkpoints/{path}.p','rb'))
layer = model['W1']
visualize_nodes(layer, f'Node Visualization for {path} (vanilla)')

path = 'save_7500'
model = pickle.load(open(f'checkpoints/{path}.p','rb'))
layer = model['W1']
visualize_nodes(layer, f'Node Visualization for {path} (split batch training)')

path = 'save_exp5_500'
model = pickle.load(open(f'checkpoints/{path}.p','rb'))
layer = model['W1']
visualize_nodes(layer, f'Node Visualization for {path} (split batch training)')

path = 'save_exp5_4000'
model = pickle.load(open(f'checkpoints/{path}.p','rb'))
layer = model['W1']
visualize_nodes(layer, f'Node Visualization for {path} (split batch training)')
```
